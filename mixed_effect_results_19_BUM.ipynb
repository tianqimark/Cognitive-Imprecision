## Load data and preprocessing

```r
data_raw <- read.table("/Users/hutianqi/Desktop/Project Cognitive Imprecision/Z1C Dataset_BU/trialsBUM_10s.csv", header=TRUE, sep=",")

data_raw$logRT <- log(1 + data_raw$RT)

# make a copy of mydata where less than 300ms and more than 10s rounds are rejected
data1 <- subset(data_raw, RT > .3 & RT < 10)

# create another dataset in wh subjects with error_choice > 5 are rejected
data2 <- subset(data1, error_choice <= 5)

```

# Split by the pre-evaluation bias

```r
# split data by whether the subject has a positive or negative pre-evaluation bias
# L: risk-lover; H: risk-hater
data1L <- subset(data1, x0 > 0)
data1H <- subset(data1, x0 < 0)

data2L <- subset(data2, x0 > 0)
data2H <- subset(data2, x0 < 0)
```

```r
# All subjects (without psychometrics)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0)):treatment + (1|subject), data1L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0)):treatment + (1|subject), data1H))
```

```r
# All subjects (without psychometrics; with interaction)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d)*scale(x0)):treatment + (1|subject), data1L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d)*scale(x0)):treatment + (1|subject), data1H))
```

```r
# without low-attention individuals (without psychometrics)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0)):treatment + (1|subject), data2L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0)):treatment + (1|subject), data2H))
```

```r
# without low-attention individuals (without psychometrics; with interaction)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d)*scale(x0)):treatment + (1|subject), data2L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d)*scale(x0)):treatment + (1|subject), data2H))
```

```r
# All subjects (with psychometrics)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0) + scale(reflection) + scale(BNT) + scale(NFC) + scale(FI)):treatment + (1|subject), data1L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0) + scale(reflection) + scale(BNT) + scale(NFC) + scale(FI)):treatment + (1|subject), data1H))
```

```r
# without low-attention individuals (with psychometrics)
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0) + scale(reflection) + scale(BNT) + scale(NFC) + scale(FI)):treatment + (1|subject), data2L))
summary(lmer(slc_lot ~ (scale(lottery_favour)*scale(WTP) + scale(d) + scale(x0) + scale(reflection) + scale(BNT) + scale(NFC) + scale(FI)):treatment + (1|subject), data2H))
```
