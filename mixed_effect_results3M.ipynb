## Load data and preprocessing

```r
data_raw <- read.table("/Users/hutianqi/Desktop/Project Cognitive Imprecision/Z1 Dataset_M/trialsAM_10s_B1.csv", header=TRUE, sep=",")

data_raw$logRT <- log(1 + data_raw$RT)

# make a copy of mydata where less than 300ms and more than 10s rounds are rejected
data1 <- subset(data_raw, RT > .3 & RT < 10)

# create another dataset in wh subjects with error_choice > 5 are rejected
data2 <- subset(data1, error_choice <= 5)

# # split the dataset by treatment
# data1_A <- subset(data1, treatment == 'A')
# data1_E <- subset(data1, treatment == 'E')

# data2_A <- subset(data2, treatment == 'A')
# data2_E <- subset(data2, treatment == 'E')
```

## 1 Predicting choice with RT and value difference alone

### 1.1 whole dataset

```r
summary(glmer(slc_lot ~ (lottery_favour + RT):treatment + (1|subject), data1, family=binomial(link = "logit")))

summary(lmer(slc_lot ~ (lottery_favour + RT):treatment + (1|subject), data1))
```

### 1.1b logRT

```r
summary(glmer(slc_lot ~ (lottery_favour + logRT):treatment + (1|subject), data1, family=binomial(link = "logit")))

summary(lmer(slc_lot ~ (lottery_favour + logRT):treatment + (1|subject), data1))
```

### 1.2 without the low-effort group

```r
summary(glmer(slc_lot ~ (lottery_favour + RT):treatment + (1|subject), data2, family=binomial(link = "logit")))

summary(lmer(slc_lot ~ (lottery_favour + RT):treatment + (1|subject), data2))
```

### 1.2b logRT

```r
summary(glmer(slc_lot ~ (lottery_favour + logRT):treatment + (1|subject), data2, family=binomial(link = "logit")))

summary(lmer(slc_lot ~ (lottery_favour + logRT):treatment + (1|subject), data2))
```
